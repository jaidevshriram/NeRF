```python
data_poses = list(map(lambda x: np.loadtxt("data/bottles/pose/"+x), os.listdir(os.path.join("data", "bottles", "pose"))))

data_poses = []

for file in sorted(os.listdir(os.path.join("data", "bottles", "pose"))):
    if file[0] == "0": # Train
        data_poses.append(np.loadtxt("data/bottles/pose/"+file))
```

# Train Poses

```python
visualizer = CameraPoseVisualizer([-3, 3], [-3, 3], [0, 3])
for i, data_pose in enumerate(data_poses):

    visualizer.extrinsic2pyramid(data_pose, plt.cm.rainbow(i / len(data_poses)), 0.3)
visualizer.colorbar(len(data_poses))
visualizer.show()
```
